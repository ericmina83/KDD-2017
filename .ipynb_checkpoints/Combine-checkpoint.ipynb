```python
import pandas as pd;
import datetime as dt;
import numpy as np;
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import scale
import pickle
from sklearn.ensemble import RandomForestRegressor

testing = pd.read_csv('testing.csv')

#hl = [100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100]
#hl = [100, 100, 100, 100]
hl = [10, 10]
reg = MLPRegressor(solver='adam', hidden_layer_sizes = hl, max_iter = 200000, learning_rate = 'adaptive', learning_rate_init = 0.01) # momentum = 0.5) #activation = 'logistic', 
rfr = RandomForestRegressor()

#reg = pickle.load(open('model_2.sav', 'rb'))
```

```python
training = pd.read_csv('training.csv')
del training['Unnamed: 0']
del training['minute']
#del training['week']
#del training['precipitation']

training_max = training.max()
training_min = training.min()
training_mean = training.mean()

training_norm = (training - training_mean) / (training_max - training_min)

X = training_norm.ix[:, :-1]
y = np.array(training_norm.ix[:, -1:]).ravel()

reg.fit(X, y)

outputfilename = "model_2.sav"

pickle.dump(reg, open(outputfilename, 'wb'))

result = reg.predict(X)

training['predict_travel_time'] = result \
    * (training_max['travel_time'] - training_min['travel_time']) + training_mean['travel_time']

training['error rate'] = abs((training['predict_travel_time'] - training['travel_time']) / training['travel_time'])

error_training = training['error rate'].mean()

#tree

rfr.fit(X, y)

training['rfr_predict_travel_time'] = rfr.predict(X) * (training_max['travel_time'] - training_min['travel_time']) + training_mean['travel_time']

training.to_csv('training_result.csv')
```

```python
testing = pd.read_csv('testing.csv')
del testing['Unnamed: 0']
del testing['minute']
#del testing['week']
#del testing['precipitation']

testing_max = testing.max()
testing_min = testing.min()
testing_ori = testing
testing_mean = testing.mean()

testing_norm = (testing - training_mean) / (training_max - training_min)
#testing_norm = (testing - testing_mean) / (testing_max - testing_min)

X = testing_norm.ix[:, :-1]

testing['predict_travel_time'] = reg.predict(X) * (training_max['travel_time'] - training_min['travel_time']) + training_mean['travel_time']
#testing['predict_travel_time'] = reg.predict(X) * (testing_max['travel_time'] - testing_min['travel_time']) + testing_mean['travel_time']

testing['error rate'] = abs(testing['predict_travel_time'] - testing['travel_time']) / testing['travel_time']

error_training = testing['error rate'].mean()

testing['rfr_predict_travel_time'] = rfr.predict(X) * (training_max['travel_time'] - training_min['travel_time']) + training_mean['travel_time']
#testing['rfr_predict_travel_time'] = rfr.predict(X) * (testing_max['travel_time'] - testing_min['travel_time']) + testing_mean['travel_time']

testing.to_csv('testing_result.csv')
```

```python
submission = pd.read_csv('submission.csv')
del submission['Unnamed: 0']
del submission['minute']
del submission['avg_travel_time']
#del submission['travel_time']
#del submission['week']
#del submission['precipitation']

#submission = submission[submission.columns[:-1]]

submission = submission.sort_values(by = ['intersection_id', 'tollgate_id', 'time_window'])

submission_max = submission.max()
submission_min = submission.min()
submission_mean = submission.mean()
submission_norm = submission.copy()

del submission_norm['time_window']

for c in submission_norm.columns:
    submission_norm[c] = (submission_norm[c] - training_mean[c]) / (training_max[c] - training_min[c])

#submission_norm = (submission - submission_mean) / (submission_max - submission_min)


X = submission_norm.ix[:, :]

result = reg.predict(X) \
    * (training_max['travel_time'] - training_min['travel_time']) + training_mean['travel_time']

submission['avg_travel_time'] = result

#submission['rfr_predict_travel_time'] = rfr.predict(X) * (training_max['travel_time'] - training_min['travel_time']) + training_mean['travel_time']

del submission['pressure']
del submission['sea_pressure']
del submission['wind_direction']
del submission['wind_speed']
del submission['temperature']
del submission['rel_humidity']
del submission['precipitation']
del submission['hour']
del submission['week']

submission.loc[submission['intersection_id'] == 1, 'intersection_id'] = 'A'
submission.loc[submission['intersection_id'] == 2, 'intersection_id'] = 'B'
submission.loc[submission['intersection_id'] == 3, 'intersection_id'] = 'C'


submission.to_csv('submission_result.csv')
```
