```python
import pandas as pd;
import datetime as dt;
import numpy as np;
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import scale
import pickle
```

```python
weather = pd.read_csv("weather (table 7)_training_update.csv")
volume = pd.read_csv("volume(table 6)_training.csv")

del volume['vehicle_type']
volume['time'] =  pd.to_datetime(volume['time'], format = "%Y-%m-%d %H:%M:%S")
volume = volume.dropna()

time = []

for c in volume['time']:
    time.append(c - dt.timedelta(seconds = c.second, minutes = c.minute))
    
volume['time'] = time
volume.to_csv("volume_time.csv")

group_dc_etc = volume.groupby(['time', 'tollgate_id', 'direction']).count().reset_index()
group = volume.groupby(['time', 'tollgate_id', 'direction', 'has_etc']).count().reset_index()

del group_dc_etc['vehicle_model']
group_dc_etc['count'] = group_dc_etc['has_etc']
del group_dc_etc['has_etc']

date = []
week = []
hour = []

for c in group_dc_etc['time']:
    date.append(c - dt.timedelta(seconds = c.second, minutes = c.minute, hours = c.hour))
    
group_dc_etc['date'] = date

group['count'] = group['vehicle_model']
del group['vehicle_model']

time_weather = []
for c in group_dc_etc['time']:
    time_weather.append(c - dt.timedelta(hours = c.hour % 3))
    
group_dc_etc['time_weather'] = time_weather

# weather

weather['time_weather'] = pd.to_timedelta(weather['hour'], unit="H") + pd.to_datetime(weather['date'], format="%Y-%m-%d")
del weather['date']
del weather['hour']
```

```python
training = pd.merge(group_dc_etc, weather, on = 'time_weather')
```

```python
del training['time_weather']
del training['date']
```
